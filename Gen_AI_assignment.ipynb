# Q2. Explain the role of probabilistic modeling in generative models.How do these models differ from discriminative models?

## Role of probabilistic modeling in generative models

Probablistic modeling plays a central role in generative models by enabling them to learn the underlying probability distribution of data. instead of focusing only on predicition, generative models aim to understand how the data itself is generated.

These models typically learn either the joint probability distribution P(X,Y), or the data distribution P(X), where X represents the input data and Y represents the class label. and hidden structures present in real-world data.

The use of probabilistic modeling allows generative models to:
 - Generative new and realistic data samples similar to the training data
 - Handle missing or incomplete data effectively
 - Model uncertainty and noise present in observations
 - Simulate complex real-world data-generation processes

Examples og generative models that  rely on probablistic modeling include Gaussian mixture models, hidden markov models, variational autoencoders, and diffusion models


## Difference between generative and discriminative models

Generative models focus on learning the entire date distrubution, including the  relationship between
input features and labels. in contrast, descriminative models learn only the conditional probability fo the label given the input.

KEY DIFFERENCES:

| Aspect                | Generative Models          | Discriminative Models                     |
| --------------------- | -------------------------- | ----------------------------------------- |
| Probability learned   | ( P(X, Y) ) or ( P(X) )    | ( P(Y \mid X) )                           |
| Objective             | Model data generation      | Classify or predict labels                |
| Data generation       | Can generate new samples   | Cannot generate data                      |
| Handling missing data | Effective                  | Limited                                   |
| Examples              | Naive Bayes, HMM, VAE, GAN | Logistic Regression, SVM, Neural Networks |


# Q3. What is the difference between autoencoders and variational autoencoders (VAE's) in the context of text generation?

| Aspect                 | Autoencoder (AE)                                            | Variational Autoencoder (VAE)                                               |
| ---------------------- | ----------------------------------------------------------- | --------------------------------------------------------------------------- |
| Latent representation  | Learns a fixed (deterministic) latent vector for each input | Learns a probability distribution (mean and variance) over latent variables |
| Latent space structure | Latent space is unstructured and may contain gaps           | Latent space is continuous, smooth, and well-structured                     |
| Sampling ability       | Random sampling usually produces meaningless outputs        | Random sampling generates coherent and meaningful text                      |
| Text generation        | Mainly used for text reconstruction, not generation         | Designed for generating new and diverse text samples                        |
| Regularization         | Uses only reconstruction loss                               | Uses reconstruction loss plus KL-divergence regularization                  |
| Uncertainty modeling   | Does not model uncertainty in latent representations        | Explicitly models uncertainty through probability distributions             |


# Q4. Describe the working of attention mechanisms in neural machine tranlation (NMT). why are they critical?

Neural mchine tranlation systems translate a source sentence into a target sentence using an encoder-decoder architecture. The attention mechanisn improves this process by allowing the  model to focus on the most relevant parts of the source sentence while generating each word of the translated output.

## Step-by-Step Working:
- 1.Encoding the source sentence
The encoder (usually an RNN, LSTM, or Transformer encoder) processes the source sentence and converts each word into a hidden state. These hidden states represent the contextual meaning of each source word.

- 2.Decoding with attention
While generating each target word, the decoder does not rely on a single fixed context vector. Instead, it:
  - Compares the current decoder state with all encoder hidden states
  - Computes attention scores indicating how relevant each source word is

- 3.Attention weight calculation
The attention scores are normalized using a softmax function to produce attention weights. These weights represent the importance of each source word for the current target word.

- 4.Context vector formation
A context vector is computed as a weighted sum of the encoder hidden states using the attention weights.

- 5.Target word generation
The decoder uses the context vector along with its current state to generate the next word in the target language.

This process is repeated for every word in the translated sentence.

## Why attention mechanisms are critical in NMT
Attention mechanisms are essential for the following reasons:

- 1.Handles long sentences effectively
Traditional encoder–decoder models compress the entire source sentence into a single vector, leading to information loss. Attention removes this limitation by dynamically accessing all source words.

- 2.Improves translation accuracy
By focusing on relevant source words, attention helps produce more accurate and fluent translations.

- 3.Enables word alignment
Attention implicitly learns alignments between source and target words, improving linguistic correctness.

- 4.Supports flexible word order
Attention allows the model to handle differences in word order between languages.

- 5.Foundation of Transformer models
Modern NMT systems like Transformers rely entirely on self-attention, making attention the core component of state-of-the-art translation models.

# Q5. What ethical considerations must be addressed when using generative AI for creative content such as poetry or storytelling?

## 1. Authorship and Ownership
- AI-generated content creates ambiguity about who owns the creative work. Since AI models are trained on large datasets created by humans, attributing authorship solely to AI may undermine the contributions of original creators. Clear guidelines are required to define ownership and credit.

## 2. Originality and Plagiarism
- Generative AI may unintentionally reproduce patterns, styles, or phrases from its training data. This raises concerns about plagiarism and intellectual property infringement, especially when generated content closely resembles existing literary works.

## 3. Copyright and Intellectual Property
- Many AI models are trained on copyrighted texts without explicit consent from authors. Using such models for creative writing can raise legal and ethical issues regarding fair use and unauthorized exploitation of creative works.

## 4. Cultural Sensitivity and Bias

- AI-generated poetry and stories may reflect cultural biases, stereotypes, or misrepresentations present in training data. This can lead to offensive or culturally insensitive content, making it important to monitor and mitigate bias.

## 5. Transparency and Disclosure

Readers should be informed when content is generated or assisted by AI. Lack of transparency can mislead audiences and erode trust in creative and literary platforms

# Q6.Use the following small text dataset to train a simple Variational Autoencoder (VAE) for text reconstruction:
- ["The sky is blue", "The sun is bright", "The grass is green",  "The night is dark", "The stars are shining"]

1. Preprocess the data (tokenize and pad the sequences).
2. Build a basic VAE model for text reconstruction.
3. Train the model and show how it reconstructs or generates similar sentences.

```python
# ===============================
# 1. IMPORT LIBRARIES
# ===============================
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences


# ===============================
# 2. DATASET
# ===============================
sentences = [
    "The sky is blue",
    "The sun is bright",
    "The grass is green",
    "The night is dark",
    "The stars are shining"
]


# ===============================
# 3. TOKENIZATION
# ===============================
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)

sequences = tokenizer.texts_to_sequences(sentences)


# ===============================
# 4. PADDING
# ===============================
max_len = max(len(seq) for seq in sequences)
padded_sequences = pad_sequences(sequences, maxlen=max_len, padding="post")


# ===============================
# 5. VOCAB SIZE
# ===============================
vocab_size = len(tokenizer.word_index) + 1


# ===============================
# 6. HYPERPARAMETERS
# ===============================
embedding_dim = 16
latent_dim = 8


# ===============================
# 7. ENCODER
# ===============================
encoder_inputs = tf.keras.Input(shape=(max_len,))
x = tf.keras.layers.Embedding(vocab_size, embedding_dim)(encoder_inputs)
x = tf.keras.layers.Flatten()(x)
x = tf.keras.layers.Dense(32, activation="relu")(x)

z_mean = tf.keras.layers.Dense(latent_dim)(x)
z_log_var = tf.keras.layers.Dense(latent_dim)(x)


def sampling(args):
    z_mean, z_log_var = args
    epsilon = tf.random.normal(shape=(tf.shape(z_mean)[0], latent_dim))
    return z_mean + tf.exp(0.5 * z_log_var) * epsilon


z = tf.keras.layers.Lambda(sampling)([z_mean, z_log_var])

encoder = tf.keras.Model(encoder_inputs, [z_mean, z_log_var, z])


# ===============================
# 8. DECODER
# ===============================
latent_inputs = tf.keras.Input(shape=(latent_dim,))
x = tf.keras.layers.Dense(32, activation="relu")(latent_inputs)
x = tf.keras.layers.Dense(max_len * vocab_size, activation="softmax")(x)
decoder_outputs = tf.keras.layers.Reshape((max_len, vocab_size))(x)

decoder = tf.keras.Model(latent_inputs, decoder_outputs)


# ===============================
# 9. VAE MODEL (CUSTOM TRAINING)
# ===============================
class VAE(tf.keras.Model):
    def __init__(self, encoder, decoder):
        super(VAE, self).__init__()
        self.encoder = encoder
        self.decoder = decoder

    def train_step(self, data):
        with tf.GradientTape() as tape:
            z_mean, z_log_var, z = self.encoder(data)
            reconstruction = self.decoder(z)

            # Reconstruction loss
            recon_loss = tf.reduce_mean(
                tf.keras.losses.sparse_categorical_crossentropy(
                    data, reconstruction
                )
            )

            # KL divergence loss
            kl_loss = -0.5 * tf.reduce_mean(
                1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var)
            )

            total_loss = recon_loss + kl_loss

        grads = tape.gradient(total_loss, self.trainable_weights)
        self.optimizer.apply_gradients(zip(grads, self.trainable_weights))

        return {
            "loss": total_loss,
            "reconstruction_loss": recon_loss,
            "kl_loss": kl_loss
        }


# ===============================
# 10. TRAIN MODEL
# ===============================
vae = VAE(encoder, decoder)
vae.compile(optimizer="adam")
vae.fit(padded_sequences, epochs=100, verbose=0)


# ===============================
# 11. RECONSTRUCTION
# ===============================
z_mean, _, z = encoder.predict(padded_sequences)
reconstructed = decoder.predict(z)
pred_tokens = np.argmax(reconstructed, axis=-1)


# ===============================
# 12. DECODE TOKENS TO TEXT
# ===============================
index_word = {v: k for k, v in tokenizer.word_index.items()}

def decode_sentence(seq):
    return " ".join(index_word.get(i, "") for i in seq if i != 0)


# ===============================
# 13. OUTPUT RESULTS
# ===============================
for original, recon in zip(padded_sequences, pred_tokens):
    print("Original     :", decode_sentence(original))
    print("Reconstructed:", decode_sentence(recon))
    print()
```

# Q8.Implement a simple attention-based encoder-decoder model for english to spanish translation using tensorflow or pytorch.

```python
import torch
import torch.nn as nn
import torch.optim as optim

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ----------------------------
# 1. Tiny toy dataset
# ----------------------------
# English -> Spanish pairs (very small, only for demo)
pairs = [
    ("hello", "hola"),
    ("how are you", "como estas"),
    ("i am fine", "estoy bien"),
    ("thank you", "gracias"),
    ("good morning", "buenos dias"),
]

# Build vocabularies
def build_vocab(sentences):
    tokens = set()
    for s in sentences:
        tokens.update(s.split())
    # add special tokens
    itos = ["<pad>", "<sos>", "<eos>"] + sorted(tokens)
    stoi = {w: i for i, w in enumerate(itos)}
    return stoi, itos

src_stoi, src_itos = build_vocab([p[0] for p in pairs])
trg_stoi, trg_itos = build_vocab([p[1] for p in pairs])

PAD_IDX = 0
SOS_IDX = 1
EOS_IDX = 2

def encode_sentence(sentence, stoi):
    ids = [stoi[w] for w in sentence.split()]
    return [SOS_IDX] + ids + [EOS_IDX]

def pad_batch(seqs, pad_idx=PAD_IDX):
    max_len = max(len(s) for s in seqs)
    padded = []
    for s in seqs:
        padded.append(s + [pad_idx] * (max_len - len(s)))
    return torch.tensor(padded, dtype=torch.long, device=device)

# Prepare training tensors (whole small dataset as one batch)
src_seqs = [encode_sentence(p[0], src_stoi) for p in pairs]
trg_seqs = [encode_sentence(p[1], trg_stoi) for p in pairs]

src_batch = pad_batch(src_seqs)  # (batch, src_len)
trg_batch = pad_batch(trg_seqs)  # (batch, trg_len)

# ----------------------------
# 2. Encoder
# ----------------------------
class Encoder(nn.Module):
    def __init__(self, input_dim, emb_dim, hid_dim):
        super().__init__()
        self.embedding = nn.Embedding(input_dim, emb_dim, padding_idx=PAD_IDX)
        self.rnn = nn.GRU(emb_dim, hid_dim, batch_first=True)

    def forward(self, src):
        # src: (batch, src_len)
        embedded = self.embedding(src)              # (batch, src_len, emb_dim)
        outputs, hidden = self.rnn(embedded)        # outputs: (batch, src_len, hid_dim)
        return outputs, hidden                      # hidden: (1, batch, hid_dim)

# ----------------------------
# 3. Attention
# ----------------------------
class Attention(nn.Module):
    def __init__(self, hid_dim):
        super().__init__()
        self.attn = nn.Linear(hid_dim * 2, hid_dim)
        self.v = nn.Linear(hid_dim, 1, bias=False)

    def forward(self, hidden, encoder_outputs, mask=None):
        # hidden: (1, batch, hid_dim)
        # encoder_outputs: (batch, src_len, hid_dim)
        batch_size = encoder_outputs.size(0)
        src_len = encoder_outputs.size(1)

        hidden = hidden.permute(1, 0, 2)           # (batch, 1, hid_dim)
        hidden = hidden.repeat(1, src_len, 1)      # (batch, src_len, hid_dim)

        energy = torch.tanh(self.attn(
            torch.cat((hidden, encoder_outputs), dim=2)
        ))                                         # (batch, src_len, hid_dim)
        scores = self.v(energy).squeeze(-1)        # (batch, src_len)

        if mask is not None:
            scores = scores.masked_fill(mask == 0, -1e9)

        attn_weights = torch.softmax(scores, dim=1)  # (batch, src_len)
        return attn_weights

# ----------------------------
# 4. Decoder with Attention
# ----------------------------
class Decoder(nn.Module):
    def __init__(self, output_dim, emb_dim, hid_dim, attention):
        super().__init__()
        self.output_dim = output_dim
        self.embedding = nn.Embedding(output_dim, emb_dim, padding_idx=PAD_IDX)
        self.rnn = nn.GRU(hid_dim + emb_dim, hid_dim, batch_first=True)
        self.fc_out = nn.Linear(hid_dim * 2 + emb_dim, output_dim)
        self.attention = attention

    def forward(self, input, hidden, encoder_outputs, mask=None):
        # input: (batch,) current token ids
        input = input.unsqueeze(1)                 # (batch, 1)
        embedded = self.embedding(input)           # (batch, 1, emb_dim)

        attn_weights = self.attention(hidden, encoder_outputs, mask)  # (batch, src_len)
        attn_weights = attn_weights.unsqueeze(1)   # (batch, 1, src_len)

        # encoder_outputs: (batch, src_len, hid_dim)
        context = torch.bmm(attn_weights, encoder_outputs)  # (batch, 1, hid_dim)

        rnn_input = torch.cat((embedded, context), dim=2)   # (batch, 1, emb_dim + hid_dim)
        output, hidden = self.rnn(rnn_input, hidden)        # output: (batch, 1, hid_dim)

        # Concatenate output, context, and embedded for final prediction
        output = output.squeeze(1)               # (batch, hid_dim)
        context = context.squeeze(1)             # (batch, hid_dim)
        embedded = embedded.squeeze(1)           # (batch, emb_dim)

        pred = self.fc_out(torch.cat((output, context, embedded), dim=1))  # (batch, output_dim)
        return pred, hidden, attn_weights.squeeze(1)

# ----------------------------
# 5. Seq2Seq wrapper
# ----------------------------
class Seq2Seq(nn.Module):
    def __init__(self, encoder, decoder):
        super().__init__()
        self.encoder = encoder
        self.decoder = decoder

    def create_mask(self, src):
        # mask: 1 for non-pad, 0 for pad
        return (src != PAD_IDX).to(src.device)

    def forward(self, src, trg, teacher_forcing_ratio=0.5):
        batch_size, trg_len = trg.shape
        trg_vocab_size = self.decoder.output_dim

        outputs = torch.zeros(batch_size, trg_len, trg_vocab_size, device=device)

        encoder_outputs, hidden = self.encoder(src)
        mask = self.create_mask(src)

        input = trg[:, 0]  # <sos>

        for t in range(1, trg_len):
            output, hidden, _ = self.decoder(input, hidden, encoder_outputs, mask)
            outputs[:, t, :] = output
            top1 = output.argmax(1)
            use_teacher = torch.rand(1).item() < teacher_forcing_ratio
            input = trg[:, t] if use_teacher else top1

        return outputs

# ----------------------------
# 6. Instantiate model
# ----------------------------
INPUT_DIM = len(src_itos)
OUTPUT_DIM = len(trg_itos)
EMB_DIM = 32
HID_DIM = 64

attn = Attention(HID_DIM)
enc = Encoder(INPUT_DIM, EMB_DIM, HID_DIM)
dec = Decoder(OUTPUT_DIM, EMB_DIM, HID_DIM, attn)

model = Seq2Seq(enc, dec).to(device)

# ----------------------------
# 7. Training (very small)
# ----------------------------
criterion = nn.CrossEntropyLoss(ignore_index=PAD_IDX)
optimizer = optim.Adam(model.parameters(), lr=0.01)

n_epochs = 200

for epoch in range(1, n_epochs + 1):
    model.train()
    optimizer.zero_grad()

    output = model(src_batch, trg_batch, teacher_forcing_ratio=0.75)
    # output: (batch, trg_len, output_dim)
    output_dim = output.shape[-1]

    output = output[:, 1:].reshape(-1, output_dim)      # skip <sos>
    trg = trg_batch[:, 1:].reshape(-1)

    loss = criterion(output, trg)
    loss.backward()
    optimizer.step()

    if epoch % 50 == 0:
        print(f"Epoch {epoch}, loss = {loss.item():.4f}")

# ----------------------------
# 8. Inference (translate a sentence)
# ----------------------------
def translate(model, sentence, max_len=10):
    model.eval()
    with torch.no_grad():
        src_seq = pad_batch([encode_sentence(sentence, src_stoi)])
        encoder_outputs, hidden = model.encoder(src_seq)
        mask = model.create_mask(src_seq)

        # start with <sos>
        input = torch.tensor([SOS_IDX], device=device)
        decoded_tokens = []

        for _ in range(max_len):
            output, hidden, attn_w = model.decoder(input, hidden, encoder_outputs, mask)
            top1 = output.argmax(1)
            if top1.item() == EOS_IDX:
                break
            decoded_tokens.append(top1.item())
            input = top1

    words = [trg_itos[idx] for idx in decoded_tokens]
    return " ".join(words)

test_sentence = "how are you"
print("Input sentence:", test_sentence)
print("Translated:", translate(model, test_sentence))
```
